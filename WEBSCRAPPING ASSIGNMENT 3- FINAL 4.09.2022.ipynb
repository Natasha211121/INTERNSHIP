2. In the above question, now scrape the following details of each product listed in first 3 pages of your 
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search 
results then scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.


```python
product_linkss=[]
start=0
end=3

for page in range(start,end):
    gt=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in gt:
        product_linkss.append(i.get_attribute('href'))
    ntb=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    ntb.click()
    time.sleep(3)
```

```python
AVB=[]

for i in product_linkss:
    driver.get(i)
    time.sleep(2)
    
    try:
        avb=driver.find_element(By.XPATH,'//div[@id="availability"]')
        AVB.append(avb.text)
    except NoSuchElementException:
        AVB.append('-')
```

```python
BRAND=[]

for p in product_linkss:
    driver.get(p)
    time.sleep(2)

    try:
        name=driver.find_elements(By.XPATH,'//span[@class="a-size-base a-color-tertiary"]')
        for i in name[0:1]:
            BRAND.append(i.text)
    except NoSuchElementException:
        BRAND.append('-')
```

```python
NAME=[]
start=0
end=3

for page in range(start,end):
    try:
        jug=driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
        for i in jug:
            NAME.append(i.text)
    except NoSuchElementException:
        NAME.append('-')
        ntb=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        ntb.click()
        time.sleep(3)
```
